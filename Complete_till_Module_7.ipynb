```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

```

#STEP 2: Load & Sort Dataset

```python
df = pd.read_csv("Smart Home Energy Consumption Optimization (1).csv")

df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values(['home_id', 'device_id', 'timestamp'])

df.head()

```

#STEP 3: Feature Engineering (Lag Feature)

```python
df['power_lag_1'] = df.groupby(['home_id', 'device_id'])['power_watt'].shift(1)
df = df.dropna()
```

#STEP 4: Select Features & Target

```python
features = [
    'power_lag_1',
    'hour_of_day',
    'day_of_week',
    'indoor_temp',
    'outdoor_temp',
    'humidity',
    'light_level',
    'price_kWh'
]

X = df[features]
y = df['power_watt']

```

#STEP 5: Time-Based Trainâ€“Test Split

```python

split = int(0.8 * len(df))

X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]
```

#STEP 6: Normalize X and y (0â€“1)

```python

X_scaler = MinMaxScaler()
y_scaler = MinMaxScaler()

X_train_scaled = X_scaler.fit_transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1,1))
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1,1))
```

#STEP 7: Train Linear Regression (Baseline)

```python

lr = LinearRegression()
lr.fit(X_train_scaled, y_train_scaled)

```

#STEP 8: Predict (0â€“1)

```python
y_pred_lr = lr.predict(X_test_scaled)
```

#STEP 9: Evaluate Baseline (0â€“1 Metrics)

```python
baseline_mae = mean_absolute_error(y_test_scaled, y_pred_lr)
baseline_rmse = mean_squared_error(y_test_scaled, y_pred_lr, squared=False)
baseline_r2 = r2_score(y_test_scaled, y_pred_lr)

print("Baseline Linear Regression (Normalized)")
print("MAE :", baseline_mae)
print("RMSE:", baseline_rmse)
print("R2  :", baseline_r2)
```

#STEP 10: Plot Actual vs Predicted (0â€“1)

```python
plt.figure(figsize=(10,5))
plt.plot(y_test_scaled[:200], label="Actual (Scaled)")
plt.plot(y_pred_lr[:200], label="Predicted (Scaled)")
plt.xlabel("Time Steps")
plt.ylabel("Normalized Power (0â€“1)")
plt.title("Baseline Linear Regression (Normalized)")
plt.legend()
plt.show()
```

**#Step 1: Create sequences**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import TimeSeriesSplit

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
```

```python
df = pd.read_csv("Smart Home Energy Consumption Optimization (1).csv")

df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values('timestamp')

# Aggregate energy per timestamp
df_energy = df.groupby("timestamp")["power_watt"].sum().reset_index()
df_energy.rename(columns={"power_watt": "total_energy"}, inplace=True)
```

```python
split = int(0.8 * len(df_energy))
train = df_energy.iloc[:split]
test  = df_energy.iloc[split:]
```

```python
scaler = MinMaxScaler()

train_scaled = scaler.fit_transform(train[["total_energy"]])
test_scaled  = scaler.transform(test[["total_energy"]])
```

```python
X_train_lr = np.arange(len(train_scaled)).reshape(-1, 1)
X_test_lr  = np.arange(len(test_scaled)).reshape(-1, 1)

lr = LinearRegression()
lr.fit(X_train_lr, train_scaled)

y_pred_lr = lr.predict(X_test_lr)

baseline_mae = mean_absolute_error(test_scaled, y_pred_lr)
baseline_mse = mean_squared_error(test_scaled, y_pred_lr)
baseline_r2  = r2_score(test_scaled, y_pred_lr)

print("Baseline Linear Regression")
print("MAE:", baseline_mae)
print("MSE:", baseline_mse)
print("R2 :", baseline_r2)
```

```python
def create_sequences(data, seq_len=24):
    X, y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i:i+seq_len])
        y.append(data[i+seq_len])
    return np.array(X), np.array(y)

X_seq, y_seq = create_sequences(train_scaled, 24)
```

```python
MAX_SAMPLES = 150_000   # reduces batches

X_seq = X_seq[:MAX_SAMPLES]
y_seq = y_seq[:MAX_SAMPLES]

X_seq = X_seq.reshape(X_seq.shape[0], X_seq.shape[1], 1)
```

```python
lstm_model = Sequential([
    LSTM(32, input_shape=(24, 1)),
    Dense(1)
])

lstm_model.compile(
    optimizer=Adam(0.001),
    loss="mse"
)

lstm_model.fit(
    X_seq,
    y_seq,
    epochs=10,
    batch_size=1024,
    verbose=1
)
```

```python
X_test_seq, y_test_seq = create_sequences(test_scaled, 24)

X_test_seq = X_test_seq.reshape(
    X_test_seq.shape[0],
    X_test_seq.shape[1],
    1
)
```

```python
y_pred_lstm = lstm_model.predict(
    X_test_seq, batch_size=1024
).flatten()

lstm_mae = mean_absolute_error(y_test_seq, y_pred_lstm)
lstm_mse = mean_squared_error(y_test_seq, y_pred_lstm)
lstm_r2  = r2_score(y_test_seq, y_pred_lstm)

print("\nLSTM Model")
print("MAE:", lstm_mae)
print("MSE:", lstm_mse)
print("R2 :", lstm_r2)
```

```python
tscv = TimeSeriesSplit(n_splits=3)

mae_scores, mse_scores, r2_scores = [], [], []

for train_idx, val_idx in tscv.split(X_seq):
    X_tr, X_val = X_seq[train_idx], X_seq[val_idx]
    y_tr, y_val = y_seq[train_idx], y_seq[val_idx]

    model = Sequential([
        LSTM(32, input_shape=(24,1)),
        Dense(1)
    ])

    model.compile(optimizer="adam", loss="mse")

    model.fit(X_tr, y_tr, epochs=3, batch_size=1024, verbose=0)

    y_pred = model.predict(X_val, batch_size=1024, verbose=0).flatten()

    mae_scores.append(mean_absolute_error(y_val, y_pred))
    mse_scores.append(mean_squared_error(y_val, y_pred))
    r2_scores.append(r2_score(y_val, y_pred))

print("\nLSTM Cross Validation")
print("MAE:", np.mean(mae_scores))
print("MSE:", np.mean(mse_scores))
print("R2 :", np.mean(r2_scores))
```

```python
tuned_model = Sequential([
    LSTM(32, return_sequences=True, input_shape=(24,1)),
    LSTM(16),
    Dense(1)
])

tuned_model.compile(
    optimizer=Adam(0.0005),
    loss="mse"
)

tuned_model.fit(
    X_seq,
    y_seq,
    epochs=15,
    batch_size=1024,
    verbose=1
)

y_pred_tuned = tuned_model.predict(
    X_test_seq, batch_size=1024
).flatten()

tuned_mae = mean_absolute_error(y_test_seq, y_pred_tuned)
tuned_mse = mean_squared_error(y_test_seq, y_pred_tuned)
tuned_r2  = r2_score(y_test_seq, y_pred_tuned)

print("\nTuned LSTM")
print("MAE:", tuned_mae)
print("MSE:", tuned_mse)
print("R2 :", tuned_r2)
```

```python
results = pd.DataFrame({
    "Model": ["Baseline", "LSTM", "LSTM CV", "Tuned LSTM"],
    "MAE": [baseline_mae, lstm_mae, np.mean(mae_scores), tuned_mae],
    "MSE": [baseline_mse, lstm_mse, np.mean(mse_scores), tuned_mse],
    "R2":  [baseline_r2, lstm_r2, np.mean(r2_scores), tuned_r2]
})

results
```

**#Dashboard and Visualization**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input
```

```python
df=pd.read_csv("Smart Home Energy Consumption Optimization (1).csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])

```

**#Build dashboard components for hourly/daily/weekly/monthly consumption.**

```python
# Hourly usage
hourly_usage = df.groupby(df['timestamp'].dt.hour)['power_watt'].sum()

# Daily usage
daily_usage = df.groupby(df['timestamp'].dt.date)['power_watt'].sum()

# Weekly usage
weekly_usage = df.groupby(df['timestamp'].dt.isocalendar().week)['power_watt'].sum()

# Monthly usage
monthly_usage = df.groupby(df['timestamp'].dt.month)['power_watt'].sum()

# Device-wise usage
device_usage = df.groupby('device_id')['power_watt'].sum()
```

**#Display device-wise usage charts using matplotlib.**

```python
# Hourly
plt.figure(figsize=(10,4))
hourly_usage.plot(kind='bar', color='skyblue')
plt.title("Hourly Energy Consumption")
plt.xlabel("Hour of Day")
plt.ylabel("Energy (Watts)")
plt.show()
```

```python
# Daily
plt.figure(figsize=(10,4))
daily_usage.plot(kind='line', color='green')
plt.title("Daily Energy Consumption")
plt.xlabel("Date")
plt.ylabel("Energy (Watts)")
plt.show()
```

```python
# Weekly
plt.figure(figsize=(10,4))
weekly_usage.plot(kind='line', color='orange')
plt.title("Weekly Energy Consumption")
plt.xlabel("Week Number")
plt.ylabel("Energy (Watts)")
plt.show()
```

```python
# Monthly
plt.figure(figsize=(10,4))
monthly_usage.plot(kind='line', color='purple')
plt.title("Monthly Energy Consumption")
plt.xlabel("Month")
plt.ylabel("Energy (Watts)")
plt.show()
```

```python
# Device-wise
plt.figure(figsize=(6,6))
device_usage.plot(kind='pie', autopct='%1.1f%%')
plt.title("Device-wise Energy Usage")
plt.show()
```

```python
# Rebuild model exactly as in training
lstm_model = Sequential([
    Input(shape=(24, 1)),
    LSTM(32, return_sequences=True),
    LSTM(16),
    Dense(1)
])

# Load trained weights
lstm_model.load_weights("tuned_lstm_energy_model_weights.h5")

# Load scaler
y_scaler = joblib.load("energy_scaler.pkl")
```

```python
def predict_energy(last_24_values):
    seq = np.array(last_24_values).reshape(1, 24, 1)
    pred_norm = lstm_model.predict(seq)
    pred_real = y_scaler.inverse_transform(pred_norm)
    return float(pred_real[0][0])
```

```python
# Take first 24 values as sample
sample_input = df['power_watt'].iloc[:24].values
predicted = predict_energy(sample_input)
print("Predicted Energy (Watts):", predicted)
```
